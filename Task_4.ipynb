```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt

# Create sample accident dataset
data = {
    'Time': ['Morning', 'Morning', 'Afternoon', 'Evening', 'Night',
             'Evening', 'Night', 'Morning', 'Afternoon', 'Evening'],
    'Weather': ['Clear', 'Rainy', 'Clear', 'Rainy', 'Fog',
                'Clear', 'Rainy', 'Fog', 'Clear', 'Clear'],
    'Accidents': [5, 8, 6, 10, 7, 9, 11, 4, 6, 8]
}

df = pd.DataFrame(data)

# Display dataset
print(df)

# Accidents by Time
df.groupby('Time')['Accidents'].sum().plot(kind='bar')
plt.title("Accidents by Time of Day")
plt.xlabel("Time of Day")
plt.ylabel("Number of Accidents")
plt.show()

# Accidents by Weather
df.groupby('Weather')['Accidents'].sum().plot(kind='bar')
plt.title("Accidents by Weather Condition")
plt.xlabel("Weather")
plt.ylabel("Number of Accidents")
plt.show()
```
